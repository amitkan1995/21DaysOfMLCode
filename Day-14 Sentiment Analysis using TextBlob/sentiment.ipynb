```python
!pip install -U textblob
!python -m textblob.download_corpora
import pandas as pd
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from textblob import TextBlob
```

```python
# Reading csv file
pd_frame = pd.read_csv('Feedback.csv', encoding='cp1252')
stop_words = set(stopwords.words('english'))
```

```python
# Having Unecessary Columns
pd_frame.drop('Unnamed: 2', axis=1, inplace=True)
```

```python
pd_frame.head(10)
```

```python
# Rearranging the order of the dataset
cols = pd_frame.columns.tolist()
cols = cols[::-1]
pd_frame = pd_frame[cols]
# Creating a column to hold value of sentiment predicted by TextBlob
pd_frame['Sentiment'] = None
```

### TextBlob.sentiment

It returns two variables - Polarity and Subjectivity
- Polarity: Ranges from [-1.0,1.0]. Postive value indicates positive sentiment. 0 means neutral sentiment and negative value means negative sentiment
- Subjectivity: Ranges from [0.0,1.0]. where 0.0 is very objective and 1.0 is very subjective. 

```python
for index, rows in pd_frame.iterrows():
    
    tokens = TextBlob(rows['Comment'])
    if tokens.sentiment[0]>0:
        rows['Sentiment'] = 'Positive'
    elif tokens.sentiment[0]==0:
        rows['Sentiment'] = 'Neutral'
    else:
        rows['Sentiment'] = 'Negative'
```
